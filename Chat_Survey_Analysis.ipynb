```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style and size for clarity
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (9,5)

# 1. Upload the CSV file
from google.colab import files
uploaded = files.upload()

# Replace with your actual file name if different
filename = list(uploaded.keys())[0]
# Use pd.read_excel for .xlsx files
df = pd.read_excel(filename)

# 2. Clean data
# Strip whitespace from both column names and data values
df.columns = df.columns.str.strip()
# Use .apply and a lambda function for cleaning, handling non-string types
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)


# 3. Preview data and missing values
print("Data preview:")
display(df.head())

print("\nMissing values per column:")
print(df.isnull().sum())

# Helper function to plot countplots with sorted order
def plot_countplot(x=None, y=None, title=None, xlabel=None, ylabel=None, palette="pastel"):
    plt.figure()
    # Order categories by frequency descending
    if x is not None:
        # Ensure the column exists before attempting to plot
        if x in df.columns:
            order = df[x].value_counts().index
            sns.countplot(x=x, data=df, order=order, palette=palette)
            plt.xlabel(xlabel if xlabel else x)
            plt.ylabel(ylabel if ylabel else "Count")
        else:
            print(f"Column '{x}' not found in the DataFrame.")
            return # Exit function if column not found
    elif y is not None:
        # Ensure the column exists before attempting to plot
        if y in df.columns:
            order = df[y].value_counts().index
            sns.countplot(y=y, data=df, order=order, palette=palette)
            plt.ylabel(ylabel if ylabel else y)
            plt.xlabel(xlabel if xlabel else "Count")
        else:
            print(f"Column '{y}' not found in the DataFrame.")
            return # Exit function if column not found
    plt.title(title)
    plt.tight_layout()
    plt.show()

# 4. Gender Distribution
plot_countplot(x='Gender',
               title='Gender Distribution',
               xlabel='Gender',
               ylabel='Number of respondents',
               palette="Set2")

# 5. Self Description
plot_countplot(y='Would you consider yourself:',
               title='How Users Consider Themselves',
               ylabel='Self Description',
               xlabel='Number of respondents',
               palette="Blues")

# 6. Vibe Match Frequency
plot_countplot(x='Do you often feel like your vibe doesn’t match with people around you?',
               title='Vibe Match Frequency',
               xlabel='Response',
               ylabel='Number of respondents',
               palette="Set1")

# 7. Preferred Connection Type
plot_countplot(y='Do you want to connect with someone who:',
               title='Preferred Connection Type',
               ylabel='Connection Preference',
               xlabel='Number of respondents',
               palette="Accent")

# 8. Vibe Preferences
plot_countplot(y='What kind of people do you usually vibe with?',
               title='People Types Users Usually Vibe With',
               ylabel='People Type',
               xlabel='Number of respondents',
               palette="coolwarm")

# 9. Emotional Understanding
plot_countplot(x='Do you feel emotionally understood by the people around you?',
               title='Feelings of Emotional Understanding',
               xlabel='Response',
               ylabel='Number of respondents',
               palette="cubehelix")

# 10. Conversation Cravings (if column exists)
if 'What do you crave more in a conversation?' in df.columns:
    plot_countplot(y='What do you crave more in a conversation?',
                   title='Conversation Cravings',
                   ylabel='Type of craving',
                   xlabel='Number of respondents',
                   palette="YlGnBu")

# 11. Interest in Anonymous Connection (if column exists)
anon_col = 'If you get a chance to anonymously connect with a like-minded student, would you use it?'
if anon_col in df.columns:
    plot_countplot(x=anon_col,
                   title='Willingness to Use Anonymous Like-Minded Connection',
                   xlabel='Response',
                   ylabel='Number of respondents',
                   palette="pastel")
```
